```python
import numpy as np
from matplotlib import pyplot as plt
import json

target = 1/np.pi
```

```python
plt.rcParams.update({'font.size': 20, 'font.family': "serif","font.serif": "CMU Serif", 'text.usetex': True})
```

```python
algos = ["QPE", "Sensing", "GMMAE", "GLSAE"]
linestyle_tuple = [
     ('dotted',                (0, (1, 2))),
     ('dashdotted',            (0, (3, 2, 1, 2))),
     ('dashed',                (0, (5, 3))),
     ('solid', 'solid'),
     ('densely dashed',        (0, (3, 1))),
     ]
fig, ax = plt.subplots(1, 2, figsize=(20,5))
for algo, (_,style) in zip(algos,linestyle_tuple):
    with open(f'./output/heisenberg/{algo.lower()}.jsonl', 'r') as json_file:
        json_list = list(json_file)
    query = []
    depth = []
    value = []
    for json_str in json_list:
        res = json.loads(json_str)
        query.append(res["query"])
        depth.append(res["depth"])
        value.append(res["value"])
    query = np.max(query,axis=0)
    depth = np.max(depth,axis=0)
    error = np.median(np.abs(np.array(value)-target), axis=0)
    ax[0].plot(depth,error,label=algo,linestyle=style,linewidth=2)
    ax[1].plot(query,error,label=algo,linestyle=style,linewidth=2)
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].set_xlabel("Depth")
ax[0].set_ylabel("Error")
ax[0].legend(loc="lower left")
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_xlabel("Queries")
ax[1].set_ylabel("Error")
ax[1].legend(loc="lower left")
plt.tight_layout()
plt.savefig("others.pdf", bbox_inches='tight')
```
